```python
import pandas as pd
import numpy as np

from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score
import matplotlib.pyplot as plt
```

```python
TRAIN = pd.read_csv('data/final_data_rev/train_meta(01234678).csv')
TEST = pd.read_csv('data/final_data_rev/test_meta(01234678).csv')
test = pd.read_csv('data/final_data_rev/test_label.csv')

X = TRAIN.drop('label', axis = 1).values
y = TRAIN['label'].values

X_train, X_val, y_train, y_val = train_test_split(X,y, test_size = 0.2, random_state = 1)
```

```python
params = {'C' : 9.991493213, 'gamma' : 0.0016548094}
model = SVC(**params, kernel = 'rbf', random_state = 1)
model.fit(X_train, y_train)
```

```python
y_predict = model.predict(X_val)
print(f1_score(y_val, y_predict, average = 'macro'))
```
